#### Requirements
Input the data source by pulling together all the tables
<br> Pivot 'New' columns and 'Existing' columns 
<br> Split the former column headers to form: 
<br> Customer Type
<br> Product
<br> Rename the measure created by the Pivot as 'Products Sold'
<br> Create a Store column from the data
<br> Remove any unnecessary data fields
<br> Turn Date into Quarter 
<br> Aggregate to form two separate outputs of the number of products sold by: (help)
<br> Product, Quarter
<br> Store, Customer Type, Product
<br> Output each data set as a csv file

<br>Solution Reference:https://github.com/kelly-gilbert/preppin-data-challenge/blob/master/2021/preppin-data-2021-03/preppin-data-2021-03.py

```python
import pandas as pd
```

```python
xl=pd.ExcelFile('./input/PD 2021 Wk 3 Input.xlsx')
```

```python
xl.sheet_names
```

```python
#loop through each sheet and concat it in the dataframe
dfIn = None
for sheet in xl.sheet_names:
    dfNew = xl.parse(sheet)
    dfNew['Store'] = sheet
    dfIn = pd.concat([dfIn, dfNew])
```

```python
dfIn.groupby('Store').max()
```

```python
# pivot the columns and rename the new value Products Sold
#https://pandas.pydata.org/docs/reference/api/pandas.melt.html?highlight=melt#pandas.melt

df = pd.melt(dfIn, id_vars=['Date','Store'], var_name='Customer-Product', value_name='Products Sold')
```

```python
df.head()
```

```python
# split the customer type and product
df[['Customer Type','Product']] = df['Customer-Product'].str.split(pat=' - ', expand=True)
```

```python
# convert date to quarter
df['Quarter'] = df['Date'].dt.quarter
```

```python
# aggregation #1: products sold by product, quarter

groupFields = ['Product', 'Quarter']
agg1 = df.groupby(groupFields)['Products Sold'].sum().reset_index()
agg1.head()
```

```python
# aggregation #2: products sold by store, customer type, product

groupFields = ['Store', 'Customer Type', 'Product']
agg2 = df.groupby(groupFields)['Products Sold'].sum().reset_index()

agg2.head()
```

```python
# output to csv

agg1.to_csv('./output/output-2021-03-product-quarter.csv', index=False)
agg2.to_csv('./output/output-2021-03-store-customertype-product.csv', index=False)
```
